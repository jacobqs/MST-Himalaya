```python
reg_calib_per = regular_ptstk_10s_5c_daily_ds.sel(time = slice('2000-01-01', '2004-12-31'))
reg_calib_per = reg_calib_per.resample(time = '1D').mean()
reg_val_per = regular_ptstk_10s_5c_daily_ds.sel(time = slice('2005-01-01', '2009-12-31'))
reg_val_per = reg_val_per.resample(time = '1D').mean()

obs_calib_per = obs_disch_ds.sel(time = slice('2000-01-01', '2004-12-30'))
obs_val_per = obs_disch_ds.sel(time = slice('2005-01-01', '2009-12-30'))
```

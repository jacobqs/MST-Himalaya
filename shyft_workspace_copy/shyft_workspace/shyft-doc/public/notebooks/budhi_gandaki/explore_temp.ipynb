```python
# Get lat and lon to extract from WFDE5
min_x = 250828
max_x = 324828
min_y = 3100180
max_y = 32031180
## Making a pyproj object for UTM Zone 45N



```

```python
ds_list = []

for i in range(len(ds.station)):
    station = ds.sel(station = i)
    if station.x.values >= min_x and station.x.values <= max_x:
        if station.y.values >= min_y and station.y.values <= max_y:
            ds_list.append(station)
```

```python
x = ds_list[1].x.values
y = ds_list[1].y.values
```

```python
crs_32645 = CRS.from_epsg(32645) # CRS from cell_data file
crs_4326 = CRS.from_epsg(4326) # Lat-lon


## Create transformer to convert from CRS to CRS

transformer = Transformer.from_crs(crs_32645, crs_4326, always_xy = True)

lon, lat = transformer.transform(x, y)
```

```python
import glob
import xarray as xr
import numpy as np
```

```python
dir_path = '/Users/jacobqs/Library/CloudStorage/OneDrive-UniversitetetiOslo/MST-HMA/Data/WFDE5'

tair_fn = glob.glob(dir_path + '/near_surface_air_temperature/*.nc')
```

```python
tair_ds = xr.open_mfdataset(tair_fn, parallel = True)
```

```python
tair_ds.head()
```

```python
# Checking for missing values
tair_ds.isnull().count()
```

```python
# Temp conversion

```

```python
# Getting daily observed temp, daily mean, daily max

t_mean = tair_ds.resample(time = 'D').mean() - 273.15 
t_min = tair_ds.resample(time = 'D').min() - 273.15 
t_max = tair_ds.resample(time = 'D').max()- 273.15 
```

```python
t_series = [t_mean, t_min, t_max]
```

```python
t_series[0].Tair.mean(dim=['time','lon', 'lat']).values
```

```python
count = []
mean = []
std = []
min = []
per_25 = []
per_50 = []
per_75 = []
max = []

for i in range(len(t_series)):
    count_val = len(t_series[i].Tair)
    mean_val = np.array(t_series[i].Tair.mean(dim = ['time', 'lon', 'lat']).values)[0]
    std_val = np.array(t_series[i].Tair.std(dim = ['time', 'lon', 'lat']).values)[0]
    min_val = np.array(t_series[i].Tair.min(dim = ['time', 'lon', 'lat']).values)[0]
    count.append(count_val)
    mean.append(mean_val)
    std.append(std_val)
    min.append(min_val)
    
```
